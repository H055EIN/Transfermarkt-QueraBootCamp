```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Make a request to the webpage
url = "https://www.transfermarkt.com/federico-dimarco/leistungsdatendetails/spieler/198116/saison//verein/0/liga/0/wettbewerb//pos/0/trainer_id/0/plus/1"
html_content = requests.get(url,headers={'User-Agent': 'Mozilla/5.0' , "Accept-Language": "en-US,en;q=0.5"}).text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find the table element using its class or id
table = soup.find_all("table")[1]
# Table headers
headers = ['Season','emp','Competition','Club','Squad','Apperance','PPG','Goals','Assists','Own_goal','Substitutions_on','Substitutions_off','Yellow_card','Second_yellow_card','Red_card','Penalty_goals','Minutes_per_goal','Minutes_played']
# Extract the table rows
rows = []
for tr in table.find_all("tr"):
    cells = []
    for td in tr.find_all("td"):
        cells.append(td.text.strip())  
    rows.append(cells)

# Convert the table data into a pandas dataframe
df = pd.DataFrame(rows, columns=headers)
```

```python
# Drop useless column and row  
df = df.drop(columns='emp')
df = df.drop(index=0)
df
```

```python
# Extrac Total Info of a player (Row #1)
total_inf = df.iloc[0, 3:]
total_inf
```
