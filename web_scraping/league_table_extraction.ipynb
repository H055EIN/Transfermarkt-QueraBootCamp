```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Make a request to the webpage
url = "https://www.transfermarkt.com/serie-a/tabelle/wettbewerb/IT1/saison_id/2021"
html_content = requests.get(url,headers={'User-Agent': 'Mozilla/5.0' , "Accept-Language": "en-US,en;q=0.5"}).text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find the table element using its class or id
table = soup.find_all("table")[1]
# Extract the table headers
headers = []
for th in table.find_all("th"):
    headers.append(th.text.strip())
headers.insert(1, 'logo')
# Extract the table rows
rows = []
for tr in table.find_all("tr"):
    cells = []
    for td in tr.find_all("td"):
        cells.append(td.text.strip())
    rows.append(cells)

# Convert the table data into a pandas dataframe
df = pd.DataFrame(rows, columns=headers)
```
